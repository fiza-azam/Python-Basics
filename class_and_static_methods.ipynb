#Instance, Class and Static Methods in Python
image.png


Instance Method



```
def method(self):
  return "Instance method called", self
```



Class Method


```
@classmethod
def classmethod(cls):
  return "class method called", cls
```



Static Method 


```
@staticmethod
def staticmethod(cls):
  return "static method called"
```



````python
class MyClass:
  def method(self):
    return "Instance method called", self

  @classmethod
  def Myclassmethod(cls):
    return "class method called", cls
  
  @staticmethod
  def Mystaticmethod():
    return "static method called"

  def Mystaticmethod2():
    return "static method called"
````

Creating object 'obj' of the class MyClass

````python
obj = MyClass()
obj.method()
````

What user writes


Static Method Example


````python
class MyNewClass:
  number_of_instance = 0

  def __init__(self):
    MyNewClass.number_of_instance = MyNewClass.number_of_instance + 1

  def printNumInstance():
    print("Number of Instances: %s"%MyNewClass.number_of_instance)
````

````python
a = MyNewClass()
b = MyNewClass()
c = MyNewClass()
d = MyNewClass()

MyNewClass.printNumInstance()
````

````python
#Call Static Method using object
a.printNumInstance()
````

````python
#Call Static Method using object with static decorator
a.printNumInstance()
````

````python
class Automobile:
  name = "default"
  def __init__(self, name):
    self.name = name
    print("This is three wheel automobile:", self.name)  
  
  @classmethod
  def twoWheels(cls, name):
    print("This is two wheel automobile:", cls.name)
  
  @staticmethod
  def size(length, width, height):
    print("Size is:",length*width*height)
    return length*width*height

autoM1 = Automobile("Truck")
autoM2 = Automobile.twoWheels("rickshaw")


autoM1.size(30, 50, 60)
Automobile.size(20, 30, 10)
````

````python
class Automobile:
  name = "default"
  def __init__(self, name, l, w, h):
    self.name = name
    print("This is three wheel automobile:", self.name, ", size:", self.size(l,w,h))  
  
  @classmethod
  def twoWheels(cls, name, l, w, h):
    print("This is two wheel automobile:", cls.name,", size:", cls.size(l,w,h))
  
  @staticmethod
  def size(length, width, height):
    print("Size is:",length*width*height)
    return length*width*height

autoM1 = Automobile("Truck", 30, 50, 60)
autoM2 = Automobile.twoWheels("rickshaw", 50, 40, 30)
#autoM1.twoWheels("test", 4, 5, 6)

#autoM1.size(30, 50, 60)
#Automobile.size(20, 30, 10)
````
